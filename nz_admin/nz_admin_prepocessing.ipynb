```python
import csv

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('NZ_Admin_JOBS.xlsx - sheet1.csv')
df
```

```python
df.set_axis(['job title', 'Col_2', 'Company','Col_4','Col_5','Col_6'], axis=1)
```

```python
df['job ID'] = df['Col_2'].apply(lambda x: x[x.index('job/')+4:x.index('?')])#extract the job id from the url

df['type'] = df['Col_2'].apply(lambda x: x[x.index('type=')+5:x.index('#')])#extract the released type from the url

df["time"] = None
df['location'] = None
df['area'] = None
df['pay'] = None

df = df.drop_duplicates(subset = ['job ID'])#drop duplicates by primary key ‘job id’

df
```

```python

for index, row in df.iterrows():

    #extract location and area and salary from 'Col_4'
    if 'area' in row['Col_4'] and ',' in row['Col_4']:
        row['area'] = row['Col_4'][row['Col_4'].index('area')+5:row['Col_4'].index(',')]
        row['location'] = row['Col_4'][row['Col_4'].index('location')+9:row['Col_4'].index('area')]
        row['pay'] =  row['Col_4'][row['Col_4'].index(',')+1:]
    elif 'area' in row['Col_4']:
        row['area'] = row['Col_4'][row['Col_4'].index('area')+5:]
        row['location'] = row['Col_4'][row['Col_4'].index('location')+9:row['Col_4'].index('area')]
    elif ',' in  row['Col_4']:
        row['location'] = row['Col_4'][row['Col_4'].index('location')+9:row['Col_4'].index(',')]
        row['pay'] =  row['Col_4'][row['Col_4'].index(',')+1:]
    else:
        row['location'] = row['Col_4'][row['Col_4'].index('location')+9:]
    #Eliminate field duplication
    if row['area']:
        i = (len(row['area'])+1)/2
        row['area'] = row['area'][:int(i)]
    if row['location']:
        i = (len(row['location'])+1)/2
        row['location'] = row['location'][:int(i)]
        
    #print(row['area'])
df
```

```python

for index, row in df.iterrows():
    if 'Private Advertiser' in row['Col_5']:
        row['Company'] = 'Private Advertiser'
```

```python
#Fill time column in days, set 0 days if less than 1 day, 1m -> 30d
for index, row in df.iterrows():
    if 'ago' in row['Col_5']:
        row['time'] = row['Col_5'][:row['Col_5'].index('ago')]
        if 'h' in row['time']:
            row['time'] = 0
        elif 'd' in row['time']:
            row['time'] = int(row['time'][:row['time'].index('d')])
        elif 'm' in row['time']:
            row['time'] = int(row['time'][:row['time'].index('m')])*30
```

```python
#extrat classification and subClassification from Col_6
df['classification'] = None
df['subClassification'] = None
for index, row in df.iterrows():
    if 'classification' in row['Col_6']:
        row['classification'] = row['Col_6'][row['Col_6'].index('classification')+15:row['Col_6'].index('subClassification')]
        row['subClassification'] = row['Col_6'][row['Col_6'].index('subClassification')+18:]
        #Eliminate field duplication
        i = (len(row['classification'])+1)/2
        row['classification'] = row['classification'][:int(i)]
        j = (len(row['subClassification'])+1)/2
        row['subClassification'] = row['subClassification'][:int(j)]
        
        
```
